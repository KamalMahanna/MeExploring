```python
import numpy as np
import plotly.graph_objects as go 
from sklearn.datasets import load_iris
```

```python
x = load_iris().data[:,0]
y = np.random.normal(0,1,len(x))
```

```python
np.std(x),np.mean(x)
```

```python
z=(x-np.mean(x))/np.std(x)
```

```python
np.std(z),np.mean(z)
```

```python
rng = np.arange(0,1,0.01)
qx = [np.quantile(z,i) for i in rng]
qy = [np.quantile(y,i) for i in rng]
```

```python
fig = go.Figure(
    [
        go.Scatter(x=qx,y=qy,mode='markers',marker=dict(size=15,opacity=0.5)),
        go.Line(x=[-3,3],y=[-3,3])
    ]
)
fig.update_layout(
    # autosize=False,
    width=600,
    height=600
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data (e.g., normal distribution)
np.random.seed(42)
data = np.random.normal(loc=5, scale=2, size=100)

# Step 1: Sort the data
sorted_data = np.sort(data)
n = len(sorted_data)

# Step 2: Compute theoretical quantiles
# Using (i - 0.5)/n as plotting position
probs = (np.arange(1, n+1) - 0.5) / n
theoretical_quantiles = np.sqrt(2) * np.erfinv(2*probs - 1)  # Inverse CDF of Normal(0,1)

# Step 3: Plot
plt.scatter(theoretical_quantiles, sorted_data, color="blue", label="Data quantiles")
plt.plot(theoretical_quantiles, np.mean(data) + np.std(data)*theoretical_quantiles, 
         color="red", label="Reference line")  # Fit line
plt.xlabel("Theoretical Quantiles (Normal)")
plt.ylabel("Sample Quantiles")
plt.title("Q-Q Plot (from scratch)")
plt.legend()
plt.show()
```
