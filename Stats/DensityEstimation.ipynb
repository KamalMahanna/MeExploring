```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=55, scale=10, size=1000)
```

```python
print(data.mean())
print(data.std())
```

```python
mu,sigma = data.mean(),data.std()
```

```python
plt.hist(data)
```

```python
def pdf_norm_dist(x, mu, sigma):
    return 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
```

```python
data_points = np.linspace(np.min(data), np.max(data), 100)
```

```python
probability_density = [pdf_norm_dist(x,mu,sigma) for x in data_points]
```
