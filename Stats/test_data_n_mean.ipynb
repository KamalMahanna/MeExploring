```python
c = np.append(np.random.normal(40,30,1000),np.random.normal(120,20,500))
sn.histplot(c,kde=True)
```

```python
data_points = np.linspace(np.min(c), np.max(c), 100)
```

```python
def kernel_gauss(x, data, bandwidth):
    def gauss_dist(x, mu, sigma):
        return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-0.5*((x-mu)/sigma)**2)
    def kde_each(x, data, bandwidth):
        return np.sum([gauss_dist(x,mu,bandwidth) for mu in data])
    
    return [kde_each(i,data,bandwidth) for i in x]
```

```python
plt.plot(kernel_gauss(data_points,c,10))
```
