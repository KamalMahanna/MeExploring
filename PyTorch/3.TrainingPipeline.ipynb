```python
x_train = torch.from_numpy(x_train).type(torch.float32)
x_test = torch.from_numpy(x_test).type(torch.float32)
y_train = torch.from_numpy(y_train).type(torch.float32)
y_test = torch.from_numpy(y_test).type(torch.float32)
```

Defining the model

```python
class MySimpleNN():
    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.weights = torch.rand((X.shape[1], 1),dtype=torch.float32, requires_grad=True)
        self.bias = torch.zeros(1, dtype=torch.float32, requires_grad=True)

    def forward_pass(self, x):
        return torch.sigmoid(torch.matmul(x, self.weights) + self.bias)
    
    def loss_fn(self,y, y_hat):
        y_hat = torch.clamp(y_hat, min=1e-7, max=1-1e-7)
        return torch.mean(-y * torch.log(y_hat) - (1-y) * torch.log(1-y_hat))
```

```python
learning_rate = 0.1
epochs = 25
```

```python
# training pipeline
model = MySimpleNN(x_train,y_train)
```

```python
# in loop of epocs
#   forward pass
#   calculate loss
#   backpropagation
#   update weights

for epoch in range(epochs):
    fp  = model.forward_pass(model.X)
    loss = model.loss_fn(model.y,fp)
    print(f"Epoch {epoch}, loss {loss}")
    loss.backward()
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad
    model.weights.grad.zero_()
    model.bias.grad.zero_()
```

```python
with torch.no_grad():
    y_pred = model.forward_pass(x_test)
    loss = model.loss_fn(y_pred, y_test)
print(f"Test Loss: {loss.item()}")
```

```python
class Model(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.linear1 = nn.Linear(num_features, 3)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(3, 1)
        self.sigmoid = nn.Sigmoid()
    def forward(self, features):
        first_layer = self.relu(self.linear1(features))
        return self.sigmoid(self.linear2(first_layer))
```

```python
# recreating

class SimpleNN(nn.Module):
    def __init__(self,num_features):
        super().__init__()
        self.linear_layer = nn.Linear(in_features=num_features,out_features=1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self,x):
        return self.sigmoid(self.linear_layer(x))
        
```

```python
lr = 0.1
epochs = 25
```

```python
loss_fn = nn.BCELoss()
model = SimpleNN(x_train.shape[1])
optimizer = torch.optim.SGD(model.parameters(), lr=lr)

for epoch in range(epochs):
    y_pred = model(x_train)
    loss = loss_fn(y_pred, y_train.view(-1,1))

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch}, Loss: {loss.item()}")
```
