```python
import numpy as np
import plotly.graph_objects as go
from collections import Counter
```

```python
class NuclearFussion:
    
    def __init__(self):
        self.NoOfNucleons = np.random.randint(1,100)
        
    
    @staticmethod
    def NuclearReaction():
        return np.random.choice(["Scatter","Absorb","Fission"],p=[0.4,0.2,0.4])
    
    @staticmethod
    def MultiplicationFactor(PreviousNoOfNucleons,NoOfNucleonsNow):
        return NoOfNucleonsNow/PreviousNoOfNucleons if PreviousNoOfNucleons!=0 else 0
    
    @staticmethod
    def NoOfNuclearFission(no):
        observations = Counter([NuclearFussion.NuclearReaction() for i in range(no)])
        return observations['Fission']
    
    
    def Simulate(self):
        
        self.NoOfNucleonsNow = self.NoOfNucleons
        for _ in range(np.random.randint(1,100)):
            self.PreviousNoOfNucleons = self.NoOfNucleonsNow
            self.NoOfNucleonsNow = NuclearFussion.NoOfNuclearFission(self.PreviousNoOfNucleons)*2
        
        return NuclearFussion.MultiplicationFactor(self.PreviousNoOfNucleons,self.NoOfNucleonsNow)
    
    def SimulateTimes(self,Frequency):
        self.AllSimulatesStorage = []
        for _ in range(Frequency):
            self.AllSimulatesStorage.append(self.Simulate())
        return self.AllSimulatesStorage
```

```python
x = NuclearFussion().SimulateTimes(100000)

fig=go.Figure(
    go.Histogram(x=x)
)
# fig.update_layout(hovemode=)
fig.show()
```
